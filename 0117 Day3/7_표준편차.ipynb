# **1. 표준편차**

* 자료의 산포도를 나타내는 수치로, 분산의 양의 제곱근으로 정의
* 표준편차가 작을수록 평균값에서 변량들의 거리가 가까움

$$\sqrt{\frac{\sum_{k=1}^N (x_i - \bar{x})^2}{n}}$$

```python
# 표준편차를 계산할 데이터의 개수를 입력하고 입력받은 개수만큼 데이터를 입력
n = int(input('표준편차를 계산할 데이터의 개수를 입력하세요 : '))
score = [0 for i in range(n)]
print(score)

for i in range(n):
  data = int(input('{}번째 데이터 : '.format(i+1)))
  score[i] = data

print(score)
```

```python
# 최대값과 최소값을 구함(정렬로 구하기)
for i in range(len(score)-1):
  for j in range(i+1, len(score)):
    if score[i] > score[j]:
      score[i], score[j] = score[j], score[i]
    
print(score)
print('최소값 : {}, 최대값 : {}'.format(score[0], score[len(score)-1]))
```

```python
# 최대값과 최소값을 제외한 나머지 데이터의 평균을 계산
total = 0
for i in range(1, len(score)-1):
  total += score[i]
mean = total / (n-2)
print('평균 : ', mean)

# mean : 
```

```python
# ∑Nk=1(xi−x¯)2 (분산)
total = 0.0
for i in range(1, len(score)-1):
  total += (score[i] - mean) ** 2
var = total / (n-2)
print(var)

# 표준편차
import math
std = math.sqrt(var)
print(std)
```

```python
print('{0}에서 최대값 {1}, 최소값 {2}을 제외한 데이터의 평균은 {3:6.4f}이고, 분산은 {4:6.4f} 표준편차는 {5:6.4f}입니다.'.format(score, score[n-1], score[0], mean, var, std))
```
