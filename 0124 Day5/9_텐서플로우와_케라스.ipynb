### **1-3. Variable**

* Variable은 변할 수 있는 상태를 저장하는데 사용되는 특별한 텐서
* 딥러닝에서는 학습해야하는 가중치(weight, bias)들을 variable로 생성

```python
tensor = tf.ones((3, 4))

# tensor[0,0] = 2  --> 실행불가
```

```python
# variable 만들기, 값 변경
variable = tf.Variable(tensor)
print(variable)

variable[0,0].assign(2)
print(variable)
```

```python
initial_value = tf.random.normal(shape=(2,2))
weight = tf.Variable(initial_value)
print(weight)
```

```python
# assign(value) : 값 설정
# assign_add(increment) : 더하기
# assign_sub(decrement) : Q빼기
new_value = tf.random.normal(shape=(2,2))
print(new_value)
weight.assign_add(new_value)
print(weight)

added_value = tf.ones(shape=(2,2))
weight.assign_sub(added_value)
print(weight)
```

### **1-4. Indexing과 Slicing**

```python
a = tf.range(1, 13)
print(a)

a =tf.reshape(a, (3,4))
print(a)
```

```python
# indexing : 인덱싱을 하게되면 차원이 감소하게 됨
print(a[1])
print(a[0, -1])
```

```python
# slicing : 슬라이싱은 차우너이 유지됨
print(a[1:-1])
print(a[:2, 2:])
```

### **1-5. Transpose(차원 바꾸기)**

```python
a = tf.range(16)
print(a)
a = tf.reshape(a, (2, 2, -1)) # -1은 자동으로 만들라는 뜻
print(a)
```

```python
print(a)
# transpose(전치할 행렬, 차원의 순서, [오퍼레이션 명칭])
b = tf.transpose(a, (2, 0, 1))  # shape를 보면 (2, 2, 4) -> (4, 2, 2)
print(b)
```

### **1-6. Tensor 연산**

```python
x = tf.constant([[1,2], [3,4]], dtype=tf.float32)
y = tf.constant([[5,6], [7,8]], dtype=tf.float32)
print(x)
print(y)
```

```python
print(tf.add(x, y))
print(tf.subtract(x, y))
print(tf.multiply(x, y))
print(tf.divide(x, y))
print(tf.matmul(x, y))
print('*' * 50)
print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x @ y)
```

```python
z = tf.range(1, 11)
print(z)
z = tf.reshape(z, (2,5))
print(z)
```

```python
# 요소의 함계
print(tf.reduce_sum(z))
```

```python
sum1 = tf.reduce_sum(z, axis=0)
print(sum1)
sum2 = tf.reduce_sum(z, axis=1)
print(sum2)
sum3 = tf.reduce_sum(z, axis=-1)
print(sum3)
```

```python
concat = tf.concat([z, z], axis=0)
print(concat)
concat = tf.concat([z, z], axis=1)
print(concat)
concat = tf.concat([z, z], axis=-1)
print(concat)
```
